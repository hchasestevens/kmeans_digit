```python
with open('data/data0', 'rb') as f:
    data0 = f.read()
with open('data/data1', 'rb') as f:
    data1 = f.read()
```

```python
import numpy as np
from scipy.misc import imresize
import itertools

zeroes = np.array_split(map(ord, data0), len(data0) / (28 * 28))
zero_imgs = [imresize(np.array_split(zero, 28), (7, 7)) for zero in zeroes]
ones = np.array_split(map(ord, data1), len(data1) / (28 * 28))
one_imgs = [imresize(np.array_split(one, 28), (7, 7)) for one in ones]
all_imgs = zero_imgs + one_imgs
all_digits = [img.flatten() for img in all_imgs]

backward_img_lookup = {
    tuple(img.flatten()): np.split(orig, 28)
    for img, orig in
    itertools.chain(zip(zero_imgs, zeroes), zip(one_imgs, ones))
}
```

```python
%matplotlib inline
from matplotlib import pyplot as pl

for i in xrange(7):
    if i == 3:
        continue
        
    zero_center_pixels = [img[3][3] for img in zero_imgs]
    zero_offcenter_pixels = [img[i][i] for img in zero_imgs]

    one_center_pixels = [img[3][3] for img in one_imgs]
    one_offcenter_pixels = [img[i][i] for img in one_imgs]

    pl.scatter(zero_center_pixels, zero_offcenter_pixels, c='r', marker='x')
    pl.scatter(one_center_pixels, one_offcenter_pixels, c='b', marker='+')
    pl.axis([0, 255, 0, 255])
    pl.title(i)
    pl.show()
```

```python
all_center_pixels = [img[3][3] for img in all_imgs]
all_offcenter_pixels = [img[4][4] for img in all_imgs]

pl.scatter(all_center_pixels, all_offcenter_pixels)
pl.axis([0, 255, 0, 255])
pl.show()
```

```python
import random
random.seed('kmeans')

num_centroids = 2

centroids = random.sample(all_digits, num_centroids)
```

```python
from itertools import count
from math import sqrt


def euclidean(point1, point2):
    return sqrt(sum(
        (a - b) ** 2
        for a, b in
        zip(point1, point2)
    ))


centroid_movements = [list() for __ in xrange(num_centroids)]
for i in count():
    centroid_point_lists = [list() for centroid in centroids]
    
    for point in all_digits:
        closest_centroid, closest_centroid_points = min(
            zip(centroids, centroid_point_lists),
            key=lambda (centroid, __): euclidean(point, centroid)
        )
        closest_centroid_points.append(point)
    
    new_centroids = [
        [sum(point[dim] for point in points) / len(points)
         for dim in 
         xrange(len(points[0]))
        ]
        for points in
        centroid_point_lists
    ]
    
    for movements, new_centroid, old_centroid in zip(centroid_movements, new_centroids, centroids):
        movements.append(euclidean(new_centroid, old_centroid))
        
    centroids = new_centroids
    
    if not any(movements[-1] for movements in centroid_movements):
        break
```

```python
all_center_pixels = [img[3][3] for img in all_imgs]
all_offcenter_pixels = [img[4][4] for img in all_imgs]

pl.scatter(all_center_pixels, all_offcenter_pixels)
pl.scatter(
    [centroid[7 * 3 + 3] for centroid in centroids], 
    [centroid[7 * 4 + 4] for centroid in centroids], 
    s=500, c='r', marker='*'
)
pl.axis([0, 255, 0, 255])
pl.show()
```

```python
a_digits, b_digits = centroid_point_lists
```

```python
a_center_pixels = [img[7 * 3 + 3] for img in a_digits]
a_offcenter_pixels = [img[7 * 4 + 4] for img in a_digits]

b_center_pixels = [img[7 * 3 + 3] for img in b_digits]
b_offcenter_pixels = [img[7 * 4 + 4] for img in b_digits]

pl.scatter(a_center_pixels, a_offcenter_pixels, c='r', marker='x')
pl.scatter(b_center_pixels, b_offcenter_pixels, c='b', marker='+')
pl.axis([0, 255, 0, 255])
pl.show()
```

```python
import matplotlib.gridspec as gridspec

num_rows = 10
num_cols = 10

fig = pl.figure(figsize=(10, 10))
gs = gridspec.GridSpec(num_rows, num_cols, wspace=0.0, hspace=0.0)

ax = [pl.subplot(gs[i]) for i in xrange(num_rows * num_cols)]
gs.update(hspace=0.0, wspace=0.0,)

for i, digit in enumerate(random.sample(a_digits, num_rows * num_cols)):
    ax[i].imshow(backward_img_lookup[tuple(digit)], 'gray')
    ax[i].axis('off')
```
